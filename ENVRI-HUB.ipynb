```python

#Search for the services in this catalogue
from SPARQLWrapper import SPARQLWrapper, JSON

sparql = SPARQLWrapper("https://linked.bodc.ac.uk/sparql/") #Choose the sparql endpoint you wish
sparql.setQuery("""

prefix dcat:<http://www.w3.org/ns/dcat#>
prefix dct:<http://purl.org/dc/terms/>

select ?catalogue ?service ?title ?servicetitle ?serviceURL  where {?catalogue  a dcat:Catalog .
?catalogue dct:title ?title .
?catalogue  dcat:service ?service .
?service dct:title ?servicetitle .
?service dcat:endpointURL ?serviceURL  .
 }

""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
#print ("Number of results: "+str(len(results["results"]["bindings"])));
for result in results["results"]["bindings"]:
    print(result["catalogue"]["value"],result["service"]["value"],result["title"]["value"],result["servicetitle"]["value"],result["serviceURL"]["value"],sep="---")
 
```

```python
#Search for stations parameters instruments lat/lons and dates for datasets

from SPARQLWrapper import SPARQLWrapper, JSON

sparql = SPARQLWrapper("https://linked.bodc.ac.uk/sparql/") #Choose the sparql endpoint you wish
sparql.setQuery("""

prefix dcat:<http://www.w3.org/ns/dcat#>

select distinct ?instrument ?parameter (?a as ?series) ?dateBegins ?lat ?lon where {


?a dcat:theme ?instrument,?parameter
filter(regex(str(?instrument),'L22') && regex(str(?parameter),'P01')).

?a <http://www.w3.org/2006/time#hasBeginning> ?dateBegins .
?a  <http://purl.org/dc/terms/spatial> ?spatial .
?spatial <http://www.w3.org/2003/01/geo/wgs84_pos#lat>     ?lat   ;                        
<http://www.w3.org/2003/01/geo/wgs84_pos#lon>     ?lon       .                      


}  limit 1000
""")
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
#print ("Number of results: "+str(len(results["results"]["bindings"])));
for result in results["results"]["bindings"]:
    print(result["series"]["value"],result["instrument"]["value"],result["parameter"]["value"],result["dateBegins"]["value"],result["lat"]["value"],result["lon"]["value"],sep="---")
 
```
